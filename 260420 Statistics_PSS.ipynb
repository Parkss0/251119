```r
#1.데이터 입력
data <- c(36.2, 36.4, 36.3, 36.1, 36.0, 36.2, 36.3, 36.1, 36.2, 36.4)

#비교할 기준값
mu_value <- 36.5

#2. 정규성 검정 (Shapiro-Wilk Test)
shapiro_result <- shapiro.test(data)
cat("\n Shapiro-wilk 정규성 검정 결과:\n")
print(shapiro_result)
```

```r
#3. One-Sample t-test (정규성을 만족할 경우)(단측 검정: mu_value보다 작은지 검정)
t_test_result <- t.test(data, mu = mu_value, alternative = "less")
cat("\nOne-Sample t-test (단측 검정) 결과:\n")
print(t_test_result)
```

```r
#4. Wilcoxon Signed-Rank Test (정규성 만족x) (단측 검정: mu_value보다 작은지 검정)
wilcox_result <- wilcox.test(data, mu = mu_value, alternative = "less")
cat("\n Wilcoxon Signed-Rank Test (단측 검정) 결과:\n")
print(wilcox_result)
```
