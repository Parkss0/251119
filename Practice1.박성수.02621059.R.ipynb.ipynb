```r
#데이터의 입력(변수명: data)

#Mean, SD, SE 구하기
data <- c(36.2, 36.4, 36.3, 36.1, 36.0, 36.2, 36.3, 36.1, 36.2, 36.4)
#평균 계산
mean_value <- mean(data)

#표준편차 계산(표본 표준편차)
sd_value <- sd(data)

#표준오차 계산(SE = SD / sqrt(n))
n <- length(data)
se_value <- sd_value / sqrt(n)

#95% 신뢰구간 계산 (t-분포 기반)
alpha <- 0.05
t_value <- qt(1 - alpha / 2, n - 1)

#신뢰구간 계산
ci_lower <- mean_value - t_value * se_value
ci_upper <- mean_value + t_value * se_value

#결과 출력
cat("평균 (Mean):", mean_value, "\n")
cat("표준편차 (Standard Deviation):", sd_value, "\n")
cat("표준오차 (Standard Error):", se_value, "\n")
cat("95% 신뢰구간 (Confidence Interval): [", ci_lower, ",", ci_upper, "]\n")
```
